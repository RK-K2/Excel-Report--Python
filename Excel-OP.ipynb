```python
import pandas as pd
import numpy as np
import matplotlib
```

```python
file_name = "sample.xlsx"
```

```python
# Reading the file
df = pd.read_excel(file_name,parse_dates=['ArrivalTimeFull'
                                         ,'SLATimeFull'],infer_datetime_format=True)
```

```python
# Split the time and Date from timestamp columns
time_stamp_column = ['ArrivalTimeFull','SLATimeFull']
for i in time_stamp_column :
    idate = 'date_' + i
    itime = 'time_' + i
    df[[idate,itime]] = df[i].astype(str).str.split(' ' ,expand=True)
```

```python
# condition to check if Arrival Time is greater than SLA time
condition1  = (df['time_ArrivalTimeFull'] > df['time_SLATimeFull'] ,df['time_ArrivalTimeFull'] < df['time_SLATimeFull'])
choice1 =['Y','N']
```

```python
# Validate the above condition and create the validation results as column 
df['SLA-Check']  = np.select(condition1,choice1,default=np.nan)
```

```python
# condition to check if 24 hour or 36 hour 
condition2  = (df['date_SLATimeFull'] == df['date_ArrivalTimeFull'] ,df['date_SLATimeFull'] != df['date_ArrivalTimeFull'])
choice2 =[24,36]
```

```python
# Validate the above condition and create the validation results as column 
df['check_24_36']  = np.select(condition2,choice2,default=np.nan)
```

```python
def sla_check_color(val):
    return ['background-color: rgb(193, 66, 66)' 
            if  v == "Y" else 'background-color: rgba(63, 191, 127, 0.8)' for  v in val]

def check_24_36_color(val):
    return ['background-color: rgba(63, 191, 127, 0.8)' 
            if  v == 24 else 'background-color: rgba(63, 191, 127, 0.8)' for  v in val]
   # return ['background-color: Yellow' if  v == 24 else 'background-color: Orange' for  v in val]
```

```python
# Pick Necessary Columns 
col=['SLA-Check' ,'check_24_36','LBD', 'Date', 'FeedID', 'GMIFD','Region','ArrivalTimeFull', 'SLATimeFull','date_ArrivalTimeFull', 'time_ArrivalTimeFull',
       'date_SLATimeFull', 'time_SLATimeFull']
```

```python
df[col].style.apply(sla_check_color, subset=['SLA-Check']).apply(check_24_36_color, subset=['check_24_36'])
```

```python
# Code to Write to the Excel along with the colour.
df[col].style.apply(sla_check_color, subset=['SLA-Check']).apply(check_24_36_color, subset=['check_24_36']).\
to_excel('SLA_CHECK.xlsx' ,engine='openpyxl',index=0)
```
